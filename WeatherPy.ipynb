```python
from citipy import citipy
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

from api_keys import weather_api_key

lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
city_lat_lng = []
cities = []
lat_indices = np.random.uniform(lat_range[0], lat_range[1], size = 1500)
lng_indices = np.random.uniform(lng_range[0], lng_range[1], size = 1500)

city_lat_lng = zip(lat_indices, lng_indices)

for x in city_lat_lng:
    city = citipy.nearest_city(x[0], x[1]).city_name
    
    if city not in cities:
        cities.append(city)

print(cities)
```

```python
#randomly generated cities
cities_list = len(cities)

print(cities_list)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
base_url =  "http://api.openweathermap.org/data/2.5/weather?"
cities_data = []
api_key = weather_api_key
record_count = 0
num_set = 1

print("Beginning Data Retrieval")
print("--------------------------------------------")
for y in cities:
    query_url = base_url + "appid=" + api_key + "&q=" + y + "&units=metric"
    city_info = requests.get(query_url).json()
    try:
        city_info = requests.get(query_url).json()
        city_name = city_info["name"]
        city_lat = city_info["coord"]["lat"]
        city_lng = city_info["coord"]["lon"]
        city_temp_max = city_info["main"]["temp_max"]
        city_humidity = city_info["main"]["humidity"]
        city_cloud = city_info["clouds"]["all"]
        city_wind_speed = city_info["wind"]["speed"]
        city_country = city_info["sys"]["country"]
        city_date = city_info["dt"]
        if record_count >=49:
            record_count = 1
            num_set += 1
            cities_data.append([city_name, city_lat, city_lng, city_temp_max, city_humidity, city_cloud, city_wind_speed, city_country, city_date])
        else:
            record_count += 1
            cities_data.append([city_name, city_lat, city_lng, city_temp_max, city_humidity, city_cloud, city_wind_speed, city_country, city_date])
        print(f"Processing Record {record_count} of Set {num_set}: {city_name}")
        
    except(KeyError):
        print("City not found")
print("Data Retrieval Complete")
print("--------------------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
cities_pd = pd.DataFrame(cities_data, columns = ["City", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Country", "Date"])
cities_pd.to_csv("Output Data/cities_weather_data.csv")

cities_pd.head()
```

## Linear Regression

```python
north_hemi_df = cities_pd.loc[cities_pd["Lat"]>=0,:]

north_hemi_df.head()
```

```python
south_hemi_df = cities_pd.loc[cities_pd["Lat"]<0,:]

south_hemi_df.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = north_hemi_df["Lat"]
y_values = north_hemi_df["Max Temp"]
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
regress_value = slope * x_values + intercept
line_eq = "y = " + str(round(slope, 2)) + " x + " + str(round(intercept, 2))

plt.scatter(x_values, y_values, edgecolors = "black", facecolors = "orange")
plt.plot(x_values, regress_value)
plt.annotate(line_eq, (10, 40), fontsize = 15, color = "red")
plt.title("Max Temp vs Latitude in Northern Hemisphere (9/23/20)")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
print("The r-squared is " + str(rvalue**2))

plt.savefig("Output Data/Max Temp vs Lat in Northern Hemisphere.png")
plt.show()
```
